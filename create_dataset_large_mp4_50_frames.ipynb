## converting to 50 frames with ffmpeg and place in npef_movienet_50f

```python
import av
import ffmpeg
```

```python
def get_info(video):
    probe = ffmpeg.probe(video)
    video_info = next(s for s in probe['streams'] if s['codec_type'] == 'video')
    width = int(video_info['width'])
    height = int(video_info['height'])
    num_frames = int(video_info['nb_frames'])
    duration = float(video_info['duration'])
    return num_frames, duration, width, height
```

```python
# transfer all study videos
origin =  "EchoNet-Dynamic/Videos/"
target = "npef_movienet_50f/"
clip_length = 50
i = 0
for index, row in result.iterrows(): #query(' EFClass != "midrange" and Split in ["TRAIN"] ').iterrows():
    i = i+1
    if i%50 == 0:
        print(i, "files copied")
       
    #print( index,row["EFClass"],row["Split"].lower() )
    file_name = row["FileName"]+".avi"
    if os.path.exists(origin + file_name):
        t_final = target+row["Split"].lower()+"/"+row["EFClass"]+"/"+row["FileName"]+".mp4"
        #print(t_final)
        
        if not os.path.exists(t_final):            
            # transform to clip_length frames mp4
            ffmpeg.input(origin+file_name).output("output.mp4").overwrite_output().run()        
            num_frames, duration, _, _ = get_info("output.mp4")
            #fps_in = num_frames / duration
            fps_out = clip_length / duration # 4 * clip_length / duration
            start_time = 0
            #end_time = duration / 4
            #end_frame = int(np.ceil(num_frames/4))
    
            (
                ffmpeg
                    .input("output.mp4")
                    #.trim(start=0, duration = duration / 4)
                    .filter('fps', fps=fps_out, round='up')
                    .output('output_hoped.mp4')
                    .overwrite_output()
                    .run()
            )
            
            num_frames_out, duration_out, _ , _ = get_info("output_hoped.mp4")
            if num_frames_out < clip_length:
                print(num_frames_out, "frames for "+ t_final)
            #input_file = ffmpeg.input("output16hoped.mp4")
            #output_file = ffmpeg.output(input_file.trim(start_frame=0, end_frame= 16), t_final')
            #ffmpeg.overwrite_output().run(output_file)
            (
                ffmpeg
                    .input("output_hoped.mp4")
                    .trim(start_frame=0, end_frame= clip_length)
                    .output(t_final)
                    .overwrite_output()
                    .run()
            )
            
            #shutil.copy(origin+file_name, t_final)        
    else: 
        print(file_name + " not in origin", origin, "not copied")
```
